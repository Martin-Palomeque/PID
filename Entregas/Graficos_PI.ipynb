```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('..\informes.mplstyle')
```

```python
path = r'..\Mediciones\Clase 3\PI\PI-0.7-0.16666666.csv'

df = pd.read_csv(path)

t = df['Time']
h = df['Position']
P = df['P']
I = df['I']
ki = df['ki'][0]

fig, ax = plt.subplots(2,1,sharex=True)

ax[0].plot(t,h)
ax[1].plot(t,P)
ax[0].set_title(f'ki = {ki}')
```
